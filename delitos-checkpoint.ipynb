```python
import geopandas as gpd
import pandas as pd
from shapely.geometry import Point

# 1. Cargar el shapefile de barrios
barrios_gdf = gpd.read_file("C:\\Users\\maria\\Desktop\\Euge\\maestria\\TT1\\barrios\\barrios.shp") 

# 2. Asegurarse de que esté en el CRS correcto
barrios_gdf = barrios_gdf.to_crs(epsg=4326)  # Lat/lon
```

```python
delitos=pd.read_excel("C:\\Users\\maria\\Downloads\\delitos_2023.xlsx")
```

```python
delitos.shape
```

```python
# 1️⃣ Corrección de barrios en delitos
barrios_excluir = [
    "0", "sin geo", "null", "no especificada", 
    "gregorio de laferrere", "banfield oeste", 
    "bernal", "dock sud", "santa maría", 
    "villa luzuriaga", "sd", "caseros", "rodrigo bueno", "nan", "florida"
]

# Normalizo todo a minúsculas para comparación
delitos["barrio"] = delitos["barrio"].str.lower().str.strip()

# Filtro barrios a excluir
delitos = delitos[~delitos["barrio"].isin(barrios_excluir)].copy()
delitos = delitos[~delitos["barrio"].isna()]

# Reemplazos específicos de errores comunes
correcciones = {
    "nunez": "nuñez",
    "contitución": "constitucion",
    "constitución": "constitucion",
    "av boedo": "boedo",
    "villa ´pueyrredon": "villa pueyrredon",
    "la boca": "boca",
    "villa general mitre": "villa gral. mitre"
}

delitos["barrio"] = delitos["barrio"].replace(correcciones)
```

```python
delitos.shape
```

```python
# Normalizar nombres en delitos
delitos["barrio"] = (
    delitos["barrio"]
    .str.lower()
    .str.normalize('NFKD').str.encode('ascii', errors='ignore').str.decode('utf-8')
    .str.strip()
)

# Normalizar nombres en barrios_gdf
barrios_gdf["nombre"] = (
    barrios_gdf["nombre"]
    .str.lower()
    .str.normalize('NFKD').str.encode('ascii', errors='ignore').str.decode('utf-8')
    .str.strip()
)

```

```python
delitos["barrio"].unique()
```

```python
barrios_gdf["nombre"].unique()
```

```python
# Agrupar delitos por barrio
delitos_total = delitos.groupby("barrio")["cantidad"].sum().reset_index(name="total_delitos")

# Merge con el GeoDataFrame de barrios
barrios_mapa = barrios_gdf.merge(
    delitos_total,
    left_on="nombre",
    right_on="barrio",
    how="left"
)

# Rellenar nulos con 0
barrios_mapa["total_delitos"] = barrios_mapa["total_delitos"].fillna(0)
barrios_mapa["centroid"] = barrios_mapa.geometry.centroid
```

```python

import matplotlib.pyplot as plt

fig, ax = plt.subplots(figsize=(14, 12))

barrios_mapa.plot(
    column="total_delitos",
    cmap="OrRd",
    linewidth=0.5,
    edgecolor="gray",
    legend=True,
    ax=ax
)

# Etiquetas con nombre y total de delitos
for idx, row in barrios_mapa.iterrows():
    plt.text(
        row["centroid"].x,
        row["centroid"].y,
        f'{row["nombre"]}\n{int(row["total_delitos"])}',
        ha='center', va='center',
        fontsize=8,
        color="black",
        bbox=dict(facecolor='white', alpha=0.5, boxstyle='round,pad=0.2')
    )
fig.savefig("delitos_por_barrio.png", dpi=300, bbox_inches="tight")
plt.title("Cantidad de Delitos por Barrio", fontsize=16)
plt.axis('off')
plt.tight_layout()
plt.show()
```

```python
import re

# Función para limpiar correctamente latitud y longitud
def limpiar_coord(valor):
    valor = str(valor)
    # Reemplazar coma decimal por punto
    valor = valor.replace(",", ".")
    # Extraer solo el primer match de - o + seguido de números y un solo punto
    match = re.search(r'-?\d+\.\d+', valor)
    if match:
        return float(match.group())
    else:
        return None  # en caso de fallo

# Aplicar la función
delitos["latitud"] = delitos["latitud"].apply(limpiar_coord)
delitos["longitud"] = delitos["longitud"].apply(limpiar_coord)
```

```python
import geopandas as gpd

# Crear GeoDataFrame con geometría de delitos
delitos_gdf = gpd.GeoDataFrame(
    delitos,
    geometry=gpd.points_from_xy(delitos["longitud"], delitos["latitud"]),
    crs="EPSG:4326"
)

# Hacer spatial join (devuelve columna 'nombre' desde barrios_gdf)
delitos_joined = gpd.sjoin(
    delitos_gdf,
    barrios_gdf[["nombre", "geometry"]],
    how="left",
    predicate="within"
)

# Renombrar la columna 'nombre' como 'barrio_corregido'
delitos_joined = delitos_joined.rename(columns={"nombre": "barrio_corregido"})
```

```python
# Si hay barrio corregido, usarlo; si no, dejar el original
delitos_joined["barrio_final"] = delitos_joined["barrio_corregido"].combine_first(delitos_joined["barrio"])
delitos_joined["barrio_final"].unique()
```

```python
# Normalizar texto para asegurar coincidencias
delitos_joined["barrio_final"] = (
    delitos_joined["barrio_final"]
    .str.lower()
    .str.normalize('NFKD').str.encode('ascii', errors='ignore').str.decode('utf-8')
    .str.strip()
)

barrios_gdf["nombre"] = (
    barrios_gdf["nombre"]
    .str.lower()
    .str.normalize('NFKD').str.encode('ascii', errors='ignore').str.decode('utf-8')
    .str.strip()
)
```

```python
# Sumar cantidad de delitos por barrio_final
delitos_total = (
    delitos_joined
    .groupby("barrio_final")["cantidad"]
    .sum()
    .reset_index(name="total_delitos")
)
```

```python
# Calcular tipo de delito más frecuente por barrio_final
tipo_mas_frecuente = (
    delitos_joined
    .groupby(["barrio_final", "tipo"])
    .size()
    .reset_index(name="cantidad")
    .sort_values(["barrio_final", "cantidad"], ascending=[True, False])
    .drop_duplicates(subset="barrio_final")
    .rename(columns={"tipo": "tipo_mas_frecuente"})
    .drop(columns="cantidad")
)
```

```python
# Hacer el merge con delitos_total para armar el dataset final
delitos_final = delitos_total.merge(
    tipo_mas_frecuente,
    on="barrio_final",
    how="left"
)

# Ver resultado
display(delitos_final)
```
