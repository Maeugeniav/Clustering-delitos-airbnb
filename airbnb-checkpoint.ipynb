## Limpiez de dataset

```python

# Convertir a string solo si aÃºn no lo es
df_airbnb["price"] = df_airbnb["price"].astype(str)

# Limpiar sÃ­mbolos $ y , y convertir a nÃºmero
df_airbnb["price"] = (
    df_airbnb["price"]
    .replace('[\$,]', '', regex=True)  # elimina $ y ,
    .replace('', np.nan)               # reemplaza strings vacÃ­os por NaN
)

# Convertir a float (coerce pone NaN donde no puede convertir)
df_airbnb["price"] = pd.to_numeric(df_airbnb["price"], errors="coerce")

# Verificar
print("Precios nulos luego de limpieza:", datos["price"].isna().sum())
```

### Importo el shapefile de barrios para hacer el spatial join

```python
import geopandas as gpd
import pandas as pd
from shapely.geometry import Point

# 1. Cargar el shapefile de barrios
barrios_gdf = gpd.read_file("C:\\Users\\maria\\Desktop\\Euge\\maestria\\TT1\\barrios\\barrios.shp") 

# 2. Asegurarse de que estÃ© en el CRS correcto
barrios_gdf = barrios_gdf.to_crs(epsg=4326)  # Lat/lon
```

```python
# 3. Crear GeoDataFrame con los datos de Airbnb
airbnb_gdf = gpd.GeoDataFrame(df_airbnb, 
                               geometry=gpd.points_from_xy(datos["longitude"], datos["latitude"]),
                               crs="EPSG:4326")

# 4. Join espacial: asignar barrio por ubicaciÃ³n
airbnb_joined = gpd.sjoin(airbnb_gdf, barrios_gdf[["nombre", "geometry"]], how="left", predicate="within")

# 5. Crear columna con el barrio corregido
airbnb_joined["barrio_corregido"] = airbnb_joined["nombre"]

# (Opcional) Comparar viejo vs nuevo
comparacion = airbnb_joined[["barrio", "barrio_corregido", "latitude", "longitude"]]
comparacion
```

```python
#Calculo la mediana de precios por barrio y tipo de habitaciÃ³n, devolviendo un DataFrame ancho (wide format) donde los tipos de habitaciÃ³n son columnas.
precios_por_tipo = (
    airbnb_joined.groupby(['barrio_corregido', 'room_type'])['price']
    .median()
    .unstack()  # convierte room_type en columnas
    .reset_index()
)
```

```python
precios_por_tipo.columns.name = None  # Elimina el nombre del Ã­ndice de columnas si existe

precios_por_tipo = precios_por_tipo.rename(columns={
    "Entire home/apt": "precio_entire_home",
    "Private room": "precio_private_room",
    "Shared room": "precio_shared_room"
})
```

```python
#Realiza un merge (join) para agregar las medianas de precio por tipo de habitaciÃ³n
# al dataframe original airbnb_joined, uniendo por la columna 'barrio_corregido'

airbnb_joined = airbnb_joined.merge(
    precios_por_tipo,
    on="barrio_corregido",
    how="left"
)
```

```python
# 1. Calcula la mediana del precio por barrio (ya imputado)
mediana_por_barrio = (
    airbnb_joined
    .groupby("barrio_corregido")["price"]
    .median()
    .reset_index()
    .rename(columns={"price": "precio_mediana_barrio"})
)

# 2. Merge con el dataframe original (airbnb_joined)
airbnb_joined = airbnb_joined.merge(
    mediana_por_barrio,
    on="barrio_corregido",
    how="left"
)
```

## Calculo  ocupacion aproximada

```python
# 1. Calcular las mÃ©tricas por barrio
ocupacion_real_barrio = airbnb_joined.groupby('barrio_corregido').agg({
    'reviews_per_month': 'mean',
    'availability_365': 'mean'
}).assign(
    ocupacion_aproximada=lambda x: x['reviews_per_month'] / x['availability_365']
)

# 2. Agregar ocupaciÃ³n mensual aproximada
ocupacion_real_barrio['ocupacion_aproximada_mes'] = ocupacion_real_barrio['ocupacion_aproximada'] * 30.44

# 3. Resetear Ã­ndice para merge
ocupacion_real_barrio_reset = ocupacion_real_barrio.reset_index()

# 4. Merge con el DataFrame original
airbnb_joined = airbnb_joined.merge(ocupacion_real_barrio_reset, on='barrio_corregido', how='left')
# Eliminar columnas duplicadas

```

```python
airbnb_joined = airbnb_joined.drop(columns=['reviews_per_month_y', 'availability_365_y'])

# Renombrar columnas originales
airbnb_joined = airbnb_joined.rename(columns={
    'reviews_per_month_x': 'reviews_per_month',
    'availability_365_x': 'availability_365'
})
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Crear copia ordenada para graficar
ocupacion_plot = airbnb_joined[['barrio_corregido', 'ocupacion_aproximada_mes']].drop_duplicates()
ocupacion_plot = ocupacion_plot.sort_values('ocupacion_aproximada_mes', ascending=False)

# Graficar
plt.figure(figsize=(14, 8))
sns.barplot(data=ocupacion_plot, x='ocupacion_aproximada_mes', y='barrio_corregido', palette='mako')

plt.title("OcupaciÃ³n mensual aproximada por barrio", fontsize=16)
plt.xlabel("OcupaciÃ³n mensual estimada (reviews por mes / disponibilidad)")
plt.ylabel("Barrio")
plt.tight_layout()
plt.show()
```

## Conteo propiedades por barrio

```python
# 1. Calcular cantidad de propiedades por barrio
cantidad_propiedades = airbnb_joined.groupby('barrio_corregido').size().reset_index(name='cantidad_propiedades')

# 2. Merge al dataset principal
airbnb_joined = airbnb_joined.merge(cantidad_propiedades, on='barrio_corregido', how='left')
```

```python
# 1. Cantidad de propiedades por barrio y tipo
conteo_tipo = airbnb_joined.groupby(['barrio_corregido', 'room_type']).size().unstack(fill_value=0)

# 2. Calcular proporciones por tipo de habitaciÃ³n
conteo_tipo_prop = conteo_tipo.div(conteo_tipo.sum(axis=1), axis=0).reset_index()

# 3. Merge al dataset principal (si lo necesitÃ¡s barrio a barrio)
airbnb_joined = airbnb_joined.merge(conteo_tipo_prop, on='barrio_corregido', how='left')
```

```python
columnas_finales =["barrio_corregido","ocupacion_aproximada","ocupacion_aproximada_mes","cantidad_propiedades","precio_entire_home","Hotel room_x","precio_private_room","precio_shared_room",'Entire home/apt',
       'Hotel room_y', 'Private room', 'Shared room']
# Crear un nuevo DataFrame con solo las columnas seleccionadas
airbnb_final = airbnb_joined[columnas_finales].copy()
```

```python
airbnb_barrio = airbnb_final.groupby("barrio_corregido").agg({          
    "ocupacion_aproximada": "first",
    "ocupacion_aproximada_mes": "first",
    "cantidad_propiedades": "first",
    "Entire home/apt": "first",
    "Hotel room_y": "first",
    "Private room": "first",
    "Shared room": "first",
    "precio_entire_home":"first",
    "Hotel room_x":"first",
    "precio_private_room":"first",
    "precio_shared_room":"first"
}).reset_index()
```
